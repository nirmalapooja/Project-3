```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Name of the CSV file
file = 'train_transaction.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
train_data = pd.read_csv(file)
```

```python
train_data.head(5)
```

```python
x_axis = train_data[train_data['card5'].notnull()]
x_axis_clean = x_axis.head(2000)
x_axis_final = x_axis_clean[['card5']]
x_axis_final
```

```python
y_axis = train_data[train_data['isFraud'].notnull()]
y_axis_clean = y_axis.head(2000)
y_axis_final = y_axis_clean[['isFraud']]
y_axis_final
```

```python
plt.scatter(x_axis_final, y_axis_final,  marker = "o",facecolors = "coral", 
              edgecolors = "black")
plt.xlim(0, 250)
plt.ylim(-1.25, 2)
plt.show()
```
